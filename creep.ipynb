```python
msk= np.random.rand(len(data2))<0.8
train=data2[msk]
test=data2[~msk]
print(msk)
print(~msk)
print(data2)
print(train)
print(test)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']])
y = np.asanyarray(train[['Rupture time (hour)']])
regr.fit (x, y)
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']])
x = np.asanyarray(test[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']])
y = np.asanyarray(test[['Rupture time (hour)']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```

```python

fig= plt.figure()
plt.plot(test[['Temperature of creep test (cel)']],regr.predict(test[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']]),color='black')
plt.scatter(test[['Temperature of creep test (cel)']],test[['Rupture time (hour)']],color='red')
plt.legend(('Pridicted', 'Real'))
plt.xlabel('Temperature of creep test (cel)')
plt.ylabel('Log(Rupture time (hour))')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn import datasets, metrics
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV


reg = GradientBoostingRegressor(n_estimators=1100, learning_rate=0.01, max_depth=3, subsample=1.0) 
                 
reg.fit(train[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']]
        ,train[['Rupture time (hour)']])       
predicted_test=reg.predict(test[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']])
predicted_train=reg.predict(train[['Ti','Al','V','Fe','C','N','H','O','Sn','Nb','Mo','Zr','Si','B','Cr','Solution treated temp(cel)','ST time (h)','Anneal temp (cel)','Annealing Time (hour)','Temperature of creep test (cel)','Stress (Mpa)','steady state strain rate (1/s)','Strain to rupture (%) (Efc)']])
R2_test=metrics.r2_score(test[['Rupture time (hour)']],predicted_test)
R2_train=metrics.r2_score(train[['Rupture time (hour)']],predicted_train)
rmse_test=np.sqrt(metrics.mean_squared_error(test[['Rupture time (hour)']],predicted_test))
rmse_train=np.sqrt(metrics.mean_squared_error(train[['Rupture time (hour)']],predicted_train))
print('R2_test:',R2_test)
print('R2_train:',R2_train)
print('rmse_test:',rmse_test)
print('rmse_train:',rmse_train)
```
